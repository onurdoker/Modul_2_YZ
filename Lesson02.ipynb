## **Dictionaries**

In this section, we will explore another data type in Python: the **dictionary**.

Dictionaries are among the data structures that make Python powerful and effective. Like lists, they are easily mutable, can grow
or shrink during runtime, and do not require copying for these operations. Additionally, a list can be inside a dictionary, and
dictionaries can be inside a list.

🧱 **Dictionary Definition**  
A dictionary is defined with comma-separated **key-value pairs** enclosed in curly braces `{}`. A colon `:` is used between each
key and its value:

```pyhton
dic = {
    <key1>: <value1>,
    <key2>: <value2>,
    ...
    <keyN>: <valueN>
}
```

For example:


````python
Movie = {"name": "Twin Peaks", "director": "David Lynch", "year": 1990}
````

Similar to different methods that we learnt to access or modify elements in a List there are methods to modify or access keys and
items in a dictionary as well.  
To access the elements in the dictionary we can use **`item()`** or **`keys()`**. Here are some examples:


````python
# keys() return all the keys in the dictionary
print("all the keys in the Movie dictionary:", Movie.keys())

# items() returns a list of tuple pairs (Keys, Value) in the dictionary
print("all the items in the Movie dictionary:", Movie.items())
````

Other examples of methods to modify a dictionary include **`get()`**, **`copy()`**, **`update()`**, **`pop()`**, etc.

## **If statement and loops**

### **🔹 If statement**

In conditional statements we use a logical outcome to decide how to proceed. Consider the following lines of code where we check
whether the value stored in the variable r is larger than a threshold value. Depending on the outcome we can proceed with
different actions i.e. we condition the next steps of our program on the value of the variable ‘r’.

The basic form of an if-statement is as follows:

```python
if condition:
    statement1 # will be done if condition is True
else:
    statement2 # will be done if condition is False
```

Note Make sure to always put " : " after condition and else.


````python
r = 2.4
if r > 2.4:
    event = 1
elif r == 2.4:
    event = 2
else:
    event = 0
print("event is ", event)
````

### **Loops**

#### **For Loops**

It often happens that you want to repeat the same analysis repetitively.  
For example you have different datasets and you want to do the analysis X on each of them. One way to do that is by using the for

Here is the basic syntax for doing that:


For example write a code to print the square of all the integers between `1` and `9`.


````python
x = [1, 2, 3]

for i in x:
    print(i)
````

````python
for i in range(1, 10, 1):
    print("square of ", i, " is ", i**2)
````

The built-in **`range(start_value, stop_value, step)`** statement returns a list of increasing or decreasing numbers.


````python
for i in range(10):
    print("square of ", i, " is ", i * i)
````

````python
my_list = [0, 3, 5, 4, 2, 7]

for i in range(len(my_list)):
    j = my_list[i]
    print("square of ", j, " is ", j * j)
````

**hint:** The Fibonacci Sequence is the series of numbers:  
0, 1, 1, 2, 3, 5, 8, 13, 21, ....  
Every next number is found by adding up the two numbers before it.


````python
# Fibonacci series up to 50
fib = [0, 1]
while True:
    nxt = fib[-1] + fib[-2]
    if nxt > 50:
        break
    fib.append(nxt)

print(fib)
````

## **Functions**

Function is a piece of code written for a specified task. For example writing a function which calculates the mean of a bunch of
values. Functions are very helpful in programming because they can be written once and be used infinite times. Learning how to
write an efficient function makes programming like playing lego; For solving any complicated problem we need to break it down to
pieces and write a function for each piece. At the end putting these functions together like lego gives us the solution to the
problem.


````python
def greeting(name):
    output = "Hello " + name
    return output


print(greeting("Jack"))
````

````python
def greeting(name):  # function header
    # docstring: explanation of what function is about. for example here:
    """Returns a greeting
    given a name."""
    # ----- body of the function -------#
    message = "Hello lovely " + name + ".\n"
    message = message + "You are the best!"
    # ----------------------------------#
    return message  # return statement


print(greeting("jack"))
````

### **Fundamental Components of a Function**

- **Function header**  
  Function begins with the keyword `def` followed by the function name and the inputs (or arguments) of the function. In the
  example above, the name of the function is greeting and the function argument is name.

- **Function docstring**  
  Function docstring is used to explain what the function does. The convention is to use `"""triple double quotes"""` around
  docstrings.

- **Function body**  
  Function body consists of what the function is supposed to do. It gets executed every time the function is called.

- **Return statement**  
  Return exits a function, optionally passing back an expression to the caller. A return statement with no arguments is the same
  as `return` `None`.

Write a function which takes two inputs 'a' and 'b' and return the sum of a and b.

2. Write a Python function to multiply all the numbers in a list.

```python
sample_list = [8, 2, 3, -1, 7]
```


````python
def multiply(sample_list):
    multiply_x = 1

    for x in sample_list:
        multiply_x = multiply_x * x
    return multiply_x


sample_list = [8, 2, 3, -1, 7]
print(multiply(sample_list))
````

3. Write a Python function that takes a list and returns a new list with unique elements of the first list.

```python
sample_list : [1,1,3,3,3,3,4,4,4,5,7,7,7]
unique_list : [1, 3, 4, 5, 7]
```


````python
def find_unique_list(list):
    copy_list = list.copy()
    unique_list = []

    for x in list:
        if copy_list.count(x) == 1:
            unique_list.append(x)
        else:
            copy_list.remove(x)

    return unique_list


sample_list = [1, 1, 3, 3, 3, 3, 4, 4, 4, 5, 7, 7, 7]
print(find_unique_list(sample_list))
````

4. Write a Python function that takes a number as a parameter and check the number is prime or not.


````python
def prime_number(n):
    if n <= 1 or type(n) != int:
        return f"{n} is not a prime number"
    else:
        for i in range(2, n):
            if n % i == 0:
                return f"{n} is not a prime number"
            else:
                return f"{n} is a prime number"


print(prime_number(11))
````
