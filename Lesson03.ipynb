### **Why is NumPy Important?**
NumPy is the fundamental library for numerical computation and multi-dimensional data (vector/matrix) operations in Python. It provides speed and efficiency, especially when working with data that is entirely or predominantly numeric.  

**1) Speed and Performance (Why "Fast"?)**
The core data structure of NumPy is the ndarray, which holds a single dtype (homogeneous), such as int32 or float64.
Due to this uniform data type:  
- It is stored more compactly in **memory** (carrying less "extra info/metadata").
- Vectorized operations and linear algebra routines generally run **much faster**.   
Consequently, many libraries (e.g., **scikit-learn**) rely heavily on NumPy arrays for data input because they align perfectly with **linear algebra** and matrix model logic.

**2) NumPy vs. Pandas: When to Use Which?**
- Math / Model Matrices / Large-Scale Numerical Computation: Use NumPy.
- Data Cleaning, Merging, Grouping, and Reporting: Use Pandas.

**3) "Table Features" Present in Pandas but Not in NumPy**
Pandas works more like "Excel":
- **Column names** and **row indices** (e.g., columns like "Age", "Gender"; row labels like "John", "Jack").
- **Join/Merge** (table concatenation).
- **GroupBy** (grouping and summarizing).
- **Time series resampling** (aggregating time series into weekly/monthly buckets).
- **Missing data handling and label alignment** (automatic alignment via index).
- **Heterogeneous tables**: Each column in a DataFrame can have a different dtype.   

NumPy is more "raw" and lower-level:
- It works with **positions** (0, 1, 2...) instead of column/row labels.
- While extremely fast, it lacks the ergonomic "table" capabilities that Pandas offers.

**4) Why is NumPy Like an "Intermediate Step"?**
NumPy is a critical bridge between standard Python lists and Pandas:
- It teaches **vector and matrix logic**.
- It is highly functional for operations like slicing, broadcasting, addition, and multiplication on multi-dimensional data (2D, 3D, etc.).

**5) Appearance: "Raw" and More Technical**
- NumPy arrays often look like nested, regular (symmetric) lists.
- Therefore, they might seem more "intimidating" at first glance compared to Pandas; however, they are incredibly powerful for numerical computation.

2.  Extract all odd numbers from an array

hint: input: `[0,1,2,3,4,5,6,7]` and desired output: `[1,3,5,7]`

````python
x = np.array([0, 1, 2, 3, 4, 5, 6, 7])
y = np.array([])

for i in x:
    if i % 2 != 0:
        y = np.append(y, i)
    else:
        continue

print(y)
````

9. Subtract the mean of each row of a matrix

````python
# 1. Reshape the array into a matrix with two rows and five columns
x = np.reshape(vector, (2, 5))
print("Mean of vector: ", np.mean(vector))

# 2. Print the mean value of each row in the matrix
row_means = np.mean(x, axis=1)
print("Row means:", row_means)

# 3.
row_means_array = np.repeat(row_means, 5).reshape(2, 5)
print("Row means array:\n", row_means_array)

# 4. Subtract the mean value of each row from the corresponding elements in the matrix
x_minus_row_mean = x - row_means_array
print(
    "Matrix with each element subtracted by its corresponding row mean:\n",
    x_minus_row_mean,
)
````
